```python
import pandas as pd
```

```python
df= pd.read_csv("marketing_campaign_uncleaned.csv",sep="\t")
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Education'].value_counts()
```

```python
df['Education'] = df['Education'].replace({'2n Cycle': '2nd Cycle'})
df['Education'].value_counts()
```

```python
df['Marital_Status'].value_counts()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'YOLO': 'Single',
    'Absurd': 'Single',
    'Alone': 'Single'
})
df['Marital_Status'].value_counts()
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
```

```python
df[['Income']].boxplot()
```

```python
df = df[df['Income'] < 600000]
```

```python
df.info()
```

```python
df.to_csv("marketing_campaign_cleaned.csv", index=False)
```
